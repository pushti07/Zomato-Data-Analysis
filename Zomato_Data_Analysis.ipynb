```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Zomato data .csv")
df
```

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)

df['rate'] = df['rate'].apply(handleRate)

df
```

```python
df.info()
```

```python
sns.countplot(x = df['listed_in(type)'] , hue=df['listed_in(type)'] , palette = 'Set1' , legend = False)
plt.xlabel('Type of Restaurants')
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
result
```

```python
plt.plot(result , c='green' , marker = 'o')
plt.xlabel("Type of Restaurant", c ='Black', size = 15)
plt.ylabel('Votes' , c = 'Black' , size = 15)
```

```python
plt.hist(df['rate'] , bins = 12 )
plt.title("Ratings Distribution")
plt.show()
```

```python
couple_Data = df['approx_cost(for two people)']
sns.countplot(x = couple_Data )
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order' , y = 'rate' , data = df)
```

```python
pivot_table = df.pivot_table(index = 'listed_in(type)' , columns = 'online_order' , aggfunc = 'size' , fill_value=0)
sns.heatmap(pivot_table , annot = True , cmap = 'YlGnBu' , fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (type)")
plt.show()
```
